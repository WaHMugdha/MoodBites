```python
# prompt: add new column
# sentiment_class = {0 : 'very bad', 1 : 'bad', 2 : 'neutral', 3 : 'good', 4: 'very good'}
def sentiment(sentiment_score):
  if sentiment_score >= -1 and sentiment_score <= -0.8:
    return 0
  elif sentiment_score > -0.8 and sentiment_score <= -0.2:
    return 1
  elif sentiment_score > -0.2 and sentiment_score <= 0.2:
    return 2
  elif sentiment_score > 0.2 and sentiment_score <= 0.8:
    return 3
  elif sentiment_score > 0.8 and sentiment_score <= 1:
    return 4



data['Sentiment_Class'] = data['Sentiment_Score'].apply(sentiment)
data.head(5)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer, PorterStemmer
import re, string
from string import punctuation

def preprocess(text):
  text = text.lower()
  text = text.strip()
  text = re.compile('<.*?>').sub('', text)
  text = re.compile('[%s]' % re.escape(string.punctuation)).sub(' ', text)
  text = re.sub('\s+', ' ', text)
  text = re.sub(r'\[[0-9]*\]',' ',text)
  text = re.sub(r'[^\w\s]', '', str(text).lower().strip())
  text = re.sub(r'\d',' ',text)
  text = re.sub(r'\s+',' ',text)
  return text

def stopword(string):
    a = [i for i in string.split() if i not in stopwords.words('english')]
    return ' '.join(a)
wl = WordNetLemmatizer()
ps = PorterStemmer()
def lemmatizer(string):
    word_pos_tags = nltk.pos_tag(word_tokenize(string))

    a = [wl.lemmatize(tag[0], get_wordnet_pos(tag[1])) for idx, tag in enumerate(word_pos_tags)]
    return " ".join(a)

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    else:
        return wordnet.NOUN

def finalpreprocess(string):
    return lemmatizer(stopword(preprocess(string)))

# finalpreprocess('A beautiful place to dine in')
```
